# 1. Carga del dataset


```python
# Completa con el código
dt_train = pd.read_csv('fashion-mnist_train.csv')
dt_test = pd.read_csv('fashion-mnist_test.csv')
```

```python
dt_train_X = dt_train.drop(columns=['label'])
dt_train_Y = dt_train['label']
dt_test_X = dt_test.drop(columns=['label'])
dt_test_Y = dt_test['label']
```
